```python
from bs4 import BeautifulSoup as BS
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from time import sleep
import csv
import pandas as pd
```

```python
driver = webdriver.Chrome()
driver.get("https://finance.yahoo.com/quote/AAPL/history/")
```

```python
page = driver.page_source
soup = BS(page, "lxml")
```

```python
content = soup.find('div', class_='table-container yf-ewueuo')
stocklist = content.find_all('div', class_='table-container svelte-ewueuo')
print(content)
```

```python
with open('AAPL_UD.csv', mode='w', newline='') as outputFile:
    AAPLCSV = csv.writer(outputFile, delimiter=',', quotechar='"', quoting = csv.QUOTE_MINIMAL)
    AAPLCSV.writerow(['date', 'high', 'low', 'close', 'adj_close', 'volume'])
```

```python
stock_list = []
for row in content.find_all('tr'):  
    cells = row.find_all('td')  
    if len(cells) >= 6:  
        dic = {
            'date': cells[0].text,
            'high': cells[1].text,
            'low': cells[2].text,
            'close': cells[3].text,
            'adj_close': cells[4].text,
            'volume': cells[5].text
        }
        stock_list.append(dic)
    else:
        print("Skipping row - not enough data cells")
```

```python
print(stock_list)
```

```python
df = pd.DataFrame(stock_list)
df.to_csv('AAPL_UD.csv', index = False)
```
